```python
html_page = requests.get('https://www.residentadvisor.net/events/us/newyork/day/2019-05-17')
soup = BeautifulSoup(html_page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
h1 = fl_col.findAll('h1')
```
